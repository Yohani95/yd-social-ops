## 1. Análisis de mensajes del Inbox

Conectamos a Supabase y exploramos los hilos de conversación.

```python
# Cargar variables de entorno (crea .env en la raíz del proyecto)
import os
from dotenv import load_dotenv
load_dotenv('../.env.local')

from supabase import create_client

url = os.getenv('NEXT_PUBLIC_SUPABASE_URL')
key = os.getenv('SUPABASE_SERVICE_ROLE_KEY')
supabase = create_client(url, key)

# Obtener hilos recientes
response = supabase.table('inbox_threads').select('*').order('updated_at', desc=True).limit(20).execute()
threads = response.data
print(f"Hilos encontrados: {len(threads)}")
```

```python
# Analizar por canal
import pandas as pd

df = pd.DataFrame(threads)
if not df.empty and 'channel' in df.columns:
    print(df['channel'].value_counts())
else:
    print("Datos de ejemplo (sin Supabase configurado):")
    df = pd.DataFrame({
        'channel': ['whatsapp', 'instagram', 'whatsapp', 'messenger'],
        'message_count': [5, 3, 12, 2]
    })
    print(df['channel'].value_counts())
```

## 2. Prueba de variantes de prompts de IA

Comparar cómo responde el bot con distintos tonos o instrucciones.

```python
# Simular llamadas a la API de IA (Groq/OpenAI/Gemini)
# En producción usarías lib/ai-providers.ts desde Node

prompts = [
    {"nombre": "Formal", "system": "Responde de forma profesional y formal."},
    {"nombre": "Amigable", "system": "Responde de forma cercana y amigable."},
    {"nombre": "Breve", "system": "Responde en máximo 2 oraciones."}
]

mensaje_usuario = "¿Cuánto cuesta el producto X?"

for p in prompts:
    print(f"\n--- {p['nombre']} ---")
    print(f"System: {p['system']}")
    print(f"User: {mensaje_usuario}")
    # Aquí llamarías a callAI([{role:'system', content: p['system']}, {role:'user', content: mensaje_usuario}])
    print("(En Jupyter con Python usarías openai/groq SDK para probar)")
```

## 3. Visualización de métricas por canal

Gráfico de barras con la distribución de conversaciones.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
canales = ['WhatsApp', 'Instagram', 'Messenger', 'TikTok']
mensajes = [45, 23, 12, 8]

plt.figure(figsize=(8, 4))
plt.bar(canales, mensajes, color=['#25D366', '#E4405F', '#0084FF', '#000000'])
plt.title('Mensajes por canal (ejemplo)')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```
